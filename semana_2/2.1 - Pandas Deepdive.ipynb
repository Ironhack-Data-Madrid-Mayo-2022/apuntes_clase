### Aplicación de funciones

````python
def suma_2(x):
    return x+2
````

````python
df.head()
````

````python
%%time

df.quantity=[suma_2(x) for x in df.quantity]

df.head()
````

````python
%%time

df.quantity=list(map(suma_2, df.quantity))

df.head()
````

````python
%%time

df.quantity=list(map(lambda x: x+2, df.quantity))

df.head()
````

````python
%%time

df.quantity=df.quantity.apply(suma_2)

df.head()
````

````python
%%time

df.quantity=df.quantity + 2

df.head()
````

````python
%%time

df.quantity=df.quantity.apply(lambda x: x+2 if x<5 else x+10)

df.head()
````

````python
%%time

df['nueva']= df.quantity / df.unitprice

df.head()
````

````python
%%time

lst=[]

for i in range(len(df)):
    
    valor=df.quantity[i] / df.unitprice[i]
    
    lst.append(valor)
    
df['nueva_2']=lst

df.head()
````
