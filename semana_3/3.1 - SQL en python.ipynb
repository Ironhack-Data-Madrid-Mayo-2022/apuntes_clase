**Primero cargar datos...**

```python
# cargar datos

import pandas as pd

df=pd.read_csv('../data/shoes.csv')

df.head()
```

```python
# seleccion de columnas

df=df[['brand', 'colors', 'sizes', 'prices.amountMax', 'imageURLs', 'ean']]

df.head()
```

```python
df=df.dropna().reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
df.colors=df.colors.apply(lambda x: x.split(','))
df.sizes=df.sizes.apply(lambda x: x.split(','))

df.head()
```

```python
# un registro por talla y color

df=df.explode('sizes').reset_index(drop=True)
df=df.explode('colors').reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
codigo=pd.read_csv('../data/upc_corpus.csv')

codigo=codigo.dropna().reset_index(drop=True)

codigo.head()
```

```python
df.ean=codigo.ean.iloc[0:len(df)]

df.head()
```

```python
df.columns=['brand', 'color', 'size', 'price', 'url', 'ean']

df.head()
```

```python
# id

df['id']=[i for i in range(len(df))]

df.head()
```

```python
# crear una base de datos

crea_db=conn.connect(host='localhost', user='root', passwd='password')

cursor=crea_db.cursor()

cursor.execute('drop database if exists productos;')
cursor.execute('create database productos;')
```

```python
# conectar a base de datos

db=conn.connect(host='localhost', user='root', passwd='password', database='productos')

cursor=db.cursor()
```

```python
# crear tabla

cursor.execute('drop table if exists zapatos;')


tabla='''

        create table zapatos(
        
            id int primary key not null,
            brand varchar(50),
            color varchar(50),
            size varchar(50),
            price float,
            url varchar(10000),
            ean varchar(50)
        );


'''


cursor.execute(tabla)
```

```python
# check 

cursor.execute('show tables;')

for x in cursor:
    print(x)
```

### Inserta datos

```python
# insert into tabla (col1, col2, col3) values (v1, v2, v3);

nombre_tabla='zapatos'

columnas=','.join(df.columns)

for i in range(len(df)):
    
    valores=tuple(df.iloc[i].values)
    
    insert_query=f'insert into {nombre_tabla} ({columnas}) values {valores};'
    
    cursor.execute(insert_query)
    
    
db.commit()           
```
